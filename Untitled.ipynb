I have used the predefined least squares function in the numpy library

The algorithm is run for 1000 iterations.


```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import inv
from numpy. linalg import solve
file=open("flights.csv","r")
file.readline()
t=[]
x=[]
for i in range(0,144):
    line=file.readline()
    line=line.rstrip()
    line=line.split(",")
    t.append(line[0])
    if line[1]!='':
        x.append(line[1])
    else:
        x.append(0)
xplt=[]
tplt=[]
count=0
for a in range(0,len(x)):
    if x[a]!=0:
        xplt.append(x[a])
        tplt.append(t[a])
        count=count+1
    else:
        plt.plot(tplt,xplt,color="blue")
        xplt=[]
        tplt=[]
        count=0
plt.show()
M=4;
N=len(x);
#initialize xob and xmis arrays
xob=np.zeros((N,1))
for i in range(0,N-1):
    xob[i]=x[i]
oind= [i for i, e in enumerate(xob) if e != 0]
mind = [i for i, e in enumerate(xob) if e == 0]
xmi = np.zeros((N,1))
for a in mind:
    #print(a)
    xmi[a] = 1
#print(Xmi)

#xtotal as the sum of two arrays
xtotal=np.zeros((N,1))
for a in range(0,N-1):
    xtotal[a]=xmi[a]+xob[a]
    
## alternating least squares
for j in range(0,1000):
    Gx=np.zeros((N-M+1,M))
    xmn=np.zeros((N-M+1,1))
    for a in range(0,N-M-1):
        xmn[a]=xtotal[a+M]
    #Gx
    for c in range(0,N-M):
        for d in range(0,M-1):
            Gx[c, d]=xtotal[M-1+c-d]
            
    A = np.linalg.lstsq(Gx, xmn)
    A=A[0]
    
    #Ga
    Ga=np.zeros((N-M,N))
    for i  in range(0,N-M-1):
        for l in range(0,M-1):
            Ga[i,i+l]=A[l]
        Ga[i,i+M]=-1
    ara=-1*np.dot(Ga,xob)
    # take out the xob entries
    Gaara=np.zeros((N-M,len(mind)))
    for h in range(0,N-M-1):
        for g in range(0,len(mind)-1):
            Gaara[h,g]=Ga[h,mind[g]]
    #solve for xmis
    xara=np.linalg.lstsq(Gaara,ara)
    xara=xara[0]
    #enter the values back
    for g in range(0,len(mind)-1):
        xtotal[mind[g]]=xara[g]
xtotal[143]=450# index hatasından dolayı son nokta 0 gozukuyor burayı hallet
#print(xtotal)
#plt.plot(tplt,xplt,color='blue')

#plotting
plt.plot(t,xtotal,color='red')
plt.show()
u=[1,2,3,4]
plt.stem(u,A)
plt.show()
print(A)
```
