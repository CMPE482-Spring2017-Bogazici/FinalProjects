$\textbf{Project 1}$

a) Reading data and ploting: 

```octave
data_sets=csvread('flights.csv');
T=data_sets(2:end,1);
X=data_sets(2:end,2);
X(X==0)=NaN;
graph1=plot(T,X);
set(graph1,'LineWidth',2);
```

Algortihm works like that:

Before starting fill missing value in $X=[X_1,X_2,..,X_N]^T$ with random number

$\textit{i)}$ obtain $G_X$ from vector $X$

$\textit{ii)}$ Calculate vector $A=[A_1,A_2,..,A_M]^T$ with this formula $A=(G_X^* G_X)^{-1}G_X^*X_{M:N}$ where $X_{M:N}=[X_{M+1}, X_{M+2},..,X_N]^T$

$\textit{iii)}$ Obtain matrix $G_A$ from vector A

$\textit{iv)}$ Calculate vector $X$ from matrix $G_A$ like that:

Assume 

$X=[X_1^{ob}, X_2^{ob}, X_3^{mi},X_4^{ob},X_5^{mi}]^T$ 

where $X_i^{ob}$s are observed index and $X_i^{mi}$s are missed index. 

And assume 

$ G_A=[\vec{a_1},\vec{a_2},\vec{a_3},\vec{a_4},\vec{a_5}] $ 

then $G_A*X=\epsilon \approx  0$
 
Define 

$G_A^{ob}=[\vec{a_1},\vec{a_2},\vec{a_4}] $ 

$G_A^{mi}=[\vec{a_3},\vec{a_5}] $

And

$X^{ob}=[X_1^{ob}, X_2^{ob}, X_4^{ob}]^T$

$X^{mi}=[X_3^{mi}, X_5^{mi}]^T$


In this scenario $G_A^{mi}X^{mi} \approx -G_A^{ob}X^{ob}$. 

Only $X^{mi}$ is unknown. Then this vector is calculated as $X^{mi}=(G_A^{mi*}G_A^{mi})^{-1}G_A^{mi*}(-G_A^{ob}X^{ob})$ and insert founded value to vactor $X$

$\textit{v)}$ Go back step $i$



$\textit{Please note that in order to avoid errors which come from machine precision, equation $(M^* M)^{-1} M^* $ is calculated from SVD. If $U \Sigma V^*=svd(M,0)$ (reduced SVD) then $(M^* M)^{-1}$ is equals to $V (\Sigma^2)^{-1} V^*$ then $(M^* M)^{-1} M^* $ is equals to $V \Sigma^{-1} U^*$}$

```octave
M=4;
missed_indexes=find(isnan(X)); %indexes of the missing datas
observed_indexes=find(1-isnan(X)); %indexes of the observed datas
NO_missing=length(missed_indexes); %number of missing datas
X_filled=X;
X_filled(isnan(X_filled))=round(rand(1,NO_missing)*600)+100; %filling missing data with 
%random integer between 100 and 700
X_observed=X;
X_observed(isnan(X_observed))=[];


NO_data=length(X_filled);
row_G_x=NO_data-M;
G_x=zeros(row_G_x,M);

%Here is iterative part. Stoppping criteria is that difference between new and previous X. 
epsilon=0.001;
X_previous=zeros(NO_data,1);
X_new=X_filled;

while(norm(X_new-X_previous)>epsilon)
    X_previous=X_new;
    
    %%%%% Obtain G_x from X %%%%%%%%
    for i=1:M 
        G_x(:,i)=X_previous(M-i+1:NO_data-i);
    end
   
    %%%%%%%%% Calculating A %%%%%%%%
    [U_G_x D_G_x V_G_x]=svd(G_x,0);
    A=V_G_x*inv(D_G_x)*U_G_x'*X_previous(M+1:end);
    
    %%%%%%%%% Obtain G_a from A %%%%%%%%
    G_a=zeros(NO_data-M,NO_data);
    row_G_a=[A(end:-1:1);-1;zeros(NO_data-M-1,1)];
    for i=1:NO_data-M
        G_a(i,:)=(circshift(row_G_a,i-1))';
    end
    
    %%%%%%%%% Calculating updated vector X %%%%%%%%
    
    %Creating G_a_observed
    G_a_observed=G_a;
    G_a_observed(:,missed_indexes)=[];
    
    %Creating G_a_missed
    G_a_missed=G_a;
    G_a_missed(:,observed_indexes)=[];
    
    %Calculating X_missed
    [U_G_a_mi D_G_a_mi V_G_a_mi]=svd(G_a_missed,0);
    X_missed=V_G_a_mi*(diag(1./diag(D_G_a_mi)))*U_G_a_mi'*(-G_a_observed*X_observed);
 
    
    %Inserting values of the X_missed into X_new
    for i=1:NO_missing
        X_new(missed_indexes(i))=X_missed(i);
    end
    
end

graph2=plot(T,X);
set(graph2,'LineWidth',3);
hold on
graph3=plot(T,X_new,'r');
hold off
figure
graph4=stem(A,'r');
set(graph4,'LineWidth',3);
```

$\textbf{Project 1-Bonus Part}$

$\epsilon_t=A_0 X_t + A_1 X_{t-1} +A_2 X_{t-2} + ... + A_M X_{t-M} + C t + D $ 

then,

$X_t =\frac{-A_1}{A_0}X_{t-1}+ \frac{-A_2}{A_0}X_{t-2}+ ... + \frac{-A_M}{A_0}X_{t-M} +\frac{-C}{A_0} t +\frac{-D}{A_0} +\frac{\epsilon_t}{A_0}$

Change of variables,

$X_t=a_1 X_{t-1} +a_2 X_{t-2} + ... + a_M X_{t-M} + c t + d + \varepsilon_t  $

Where data vector $X=[X_1, X_2, ... X_N]^T$ and $N$ denotes number of the total datas.

If datas are fully known, below equation can be used in order to determine coefficients:

$\begin{bmatrix} 
X_{M+1} \\ 
X_{M+2}\\ 
.\\
.\\
.\\
X_N
\end{bmatrix}= \begin{bmatrix}
X_M & X_{M-1} & . & . & . & X_1 & M+1 & 1\\ 
X_{M+1} & X_{M} & . & . & . & X_2 & M+2 & 1\\ 
. & . & . &  &  & .& . & .\\ 
. & . &  & . &  & .& . & .\\ 
. &  .&  &  & . & .& . & .\\ 
X_{N-1} & X_{N-2} & . & . & . &X_{N-M} & N & 1
\end{bmatrix}.\begin{bmatrix}
a_1\\ 
a_2\\ 
.\\
.\\
.\\
a_M\\
c\\
d
\end{bmatrix}+\begin{bmatrix}
\varepsilon_{M+1}\\ 
\varepsilon_{M+2}\\ 
.\\ 
.\\ 
.\\
\varepsilon_{N}
\end{bmatrix}$

It is writen as $X_{M+1:N}=\mathbb{X} a + \varepsilon$ and solution is $a=(\mathbb{X}^*\mathbb{X})^{-1} \mathbb{X}^* X_{M+1:N}$

If coefficients are fully known, below equation can be used in order to determine datas:

$-c\begin{bmatrix}
M+1\\ 
M+2\\ 
.\\ 
.\\ 
.\\
N
\end{bmatrix}-d=\begin{bmatrix}
a_M & a_{M-1} & . & . & a_1 & -1 &0 &.&.&0\\ 
0 & a_M & a_{M-1} & . & . &a_1 & -1 & . & . & 0\\ 
. & . & . & . &  & &  &  &  &.\\ 
. & . & . &  &.  & &  &  &  &.\\ 
. & . & . &  &  &.&  &  &  &.\\ 
0 & 0 & . & . & a_M & a_{M-1} & .&. &a_1&-1
\end{bmatrix}\begin{bmatrix}
X_1\\ 
X_2\\ 
.\\ 
.\\ 
.\\ 
X_N
\end{bmatrix}+\begin{bmatrix}
\varepsilon_{M+1}\\ 
\varepsilon_{M+2}\\ 
.\\ 
.\\ 
.\\
\varepsilon_{N}
\end{bmatrix}$

This matrix equation can be formulated as $\varsigma =A X + \varepsilon$.

Missed parameter in vector $X$ can be found as previous question.

```octave
M=5;
row_G_x_B=NO_data-M;
G_x_B=zeros(row_G_x_B,M+2);

X_previous_B=zeros(NO_data,1);
X_new_B=X_filled;

while(norm(X_new_B-X_previous_B)>epsilon)
    X_previous_B=X_new_B;
    
    %%%%% Obtain G_x_B from X %%%%%%%%
    for i=1:M 
        G_x_B(:,i)=X_previous_B(M-i+1:NO_data-i);
    end
    G_x_B(:,M+1)=M+1:NO_data;
    G_x_B(:,M+2)=1;
    
    %%%%%%%%% Calculating A_B %%%%%%%%
    [U_G_x_B D_G_x_B V_G_x_B]=svd(G_x_B,0);
    A_B=V_G_x_B*inv(D_G_x_B)*U_G_x_B'*X_previous_B(M+1:end);
    
    %%%%%%%%% Obtain G_a_B from A_B %%%%%%%%
    G_a_B=zeros(NO_data-M,NO_data);
    row_G_a_B=[A_B(end-2:-1:1);-1;zeros(NO_data-M-1,1)];
    for i=1:NO_data-M
        G_a_B(i,:)=(circshift(row_G_a_B,i-1))';
    end
    
    %%%%%%%%% Calculating updated vector X %%%%%%%%
    
    %Creating G_a_B_observed
    G_a_B_observed=G_a_B;
    G_a_B_observed(:,missed_indexes)=[];
    
    %Creating G_a_B_missed
    G_a_B_missed=G_a_B;
    G_a_B_missed(:,observed_indexes)=[];
    
    %Calculating X_missed_B
    [U_G_a_B_mi D_G_a_B_mi V_G_a_B_mi]=svd(G_a_B_missed,0);
    X_missed_B=V_G_a_B_mi*(diag(1./diag(D_G_a_B_mi)))*U_G_a_B_mi'*(-(G_a_B_observed*X_observed+A_B(M+1)*[M+1:NO_data]'+A_B(M+2)));
    
    %Inserting values of the X_missed_B into X_new_B
    for i=1:NO_missing
        X_new_B(missed_indexes(i))=X_missed_B(i);
    end
    
end


hold on
graph5=plot(T,X);
set(graph5,'LineWidth',3);
graph6=plot(T,X_new_B,'r');
title("Bonus part and Original Data");
hold off
figure;
hold on
graph7=plot(T,X);
set(graph7,'LineWidth',5);

graph8=plot(T,X_new_B,'r');
set(graph8,'LineWidth',2);

graph9=plot(T,X_new,'k');
set(graph9,'LineWidth',2);
legend('original data','bonus part','part 1');
hold off

```

$\textbf{Project 2}$

a) Reading image and ploting: 

```octave
X_im=imread("stones_rgb.jpg");
imshow(X_im);

% Creating vector C
h=64;
C=zeros(3,h);
for x=1:h
    for y=1:h
        C(:,x+h*(y-1))=[X_im(x,y,1);X_im(x,y,2);X_im(x,y,3)];
    end
end
```

Since, calculating adjacency matrix $W$ takes long time, this calculation should be optimized. All diagonal elements of the $W$ is zero and $W_{ij}=W_{ji}$, only calculating upper part of the matrix is enough.



```octave
% Creating adjacency matrix W
norm_weight=[.299 .587 .114;-.14713 -.28886 .436;.615 -.51499 -.10001];
r=2;
sigma=10;
W=zeros(h*h,h*h);
for i=1:h*h
    for j=i+1:h*h
        x_i=mod(mod(i,h)+h-1,h)+1; %in here mod(64k,64)=64 and rest is same with mod operator
        x_j=mod(mod(j,h)+h-1,h)+1;
        y_i=(i-x_i)/64+1;
        y_j=(j-x_j)/64+1;
        if (abs(x_i-x_j)<=r) && (abs(y_i-y_j)<=r)
            W(i,j)=exp( -( (C(:,i)-C(:,j))'*norm_weight*(C(:,i)-C(:,j))) /(2*sigma*sigma) );
            W(j,i)=W(i,j);  %weighted norm square (||x||_W)^2 is calculated as x'*W*x 
        end
    end
end
```

Here is Laplacian matrix $L$

```octave
% Creating diagonal degree matrix D
D=zeros(h*h,h*h);
for i=1:h*h
    D(i,i)=sum(W(i,:));
end
% Creating Laplacian matrix L
L=diag(diag(D).^(-1/2))*W*diag(diag(D).^(-1/2));
```

Since matrix $L$ is symmetric, below algorithm can be used for calculating first $k^{th}$ eigenvalues which have biggest magnitude, and corresponding eigenvectors. 

Diagonal elements of the vector $R$ are eigenvalues and columns of the vector $Q$ are corresponding eigenvectors.

```octave
k=5;
[Q,R]=qr(rand(h*h,k),0);
Q_old=zeros(h*h,1);
eps=1/4096;
while norm(Q-Q_old)>eps
    Q_old=Q;
    Z=L*Q;
    [Q,R]=qr(Z,0);
end
R
```

Obtaining matrix $V$

```octave
V=diag(diag(D).^(-0.5))*Q;
```

In below cell, histograms and clustered images of the each columns of the matrix $V$ are shown. 

One of the reshaped column of matrix "$V$" is "$im$" (it is turned to 64x64 from 4096x1). If an element of the "$im$" is negative, corresponding pixel in original image is assigned to cluster one. Otherwise, pixel is assigned to cluster two.

```octave
for t=1:5
    im=reshape(V(:,t),64,64);
    imV=zeros(64,64,3); 
    imV(:,:,1)=im;
    cl1=zeros(64,64,3);
    cl2=zeros(64,64,3);
    dum1=im;
    dum1(dum1>0)=0;
    dum1(dum1<0)=1;
    for i=1:3
        cl1(:,:,i)=dum1.*X_im(:,:,i);
    end
    cl1= cast(cl1,"uint8");
    dum2=im;
    dum2(dum2>0)=1;
    dum2(dum2<0)=0;
    
    for i=1:3
        cl2(:,:,i)=dum2.*X_im(:,:,i);
        imV(:,:,i)=im*4096;
    end
    cl2= cast(cl2,"uint8");
    imV= cast(imV,"uint8"); 
    figure;
    subplot(2,2,1);
    hist(V(:,k));
    title(['histogram of V_'  num2str(t)]);
    subplot(2,2,2);
    imshow(imV); 
    title(['V_'  num2str(t)]);
    subplot(2,2,3);
    imshow(cl1);
    title("Cluster 1");
    subplot(2,2,4);
    imshow(cl2);
    title("Cluster 2");
end
```

$\textbf{Project 2 - Bonus Part}$

In this part, every row of the $V$ matrix corresponds to coordinate of the regarding pixel. With the information of the location of the each pixel, K-means algorithm is used to cluster pixels.

Algorithm works like that:

$\textit{i)}$ Normalized rows of the matrix $V$

$\textit{ii)}$ Randomly produce K centers and also normalize them.

$\textit{iii)}$ Assign each pixels to nearest center

$\textit{iv)}$ Update centers (take average of the pixels' coordinate in corresponding cluster)

$\textit{v)}$ If there is no point arround a center or there are less than $min\_pixel$ pixel, randomly create new coordinate, normalize and update center's coordinate.

$\textit{vi)}$ Turn back step $\textit{iii}$

```octave
number_of_cluster=16;
min_pixel=10;

% Normalizing rows of the V
V_norm=V./sqrt(sum(V.^2,2));

row_len=length(V(1,:));
centers_of_cluster=rand(number_of_cluster,row_len)-0.5; %creating cluster center randomly between 0.5 and -0.5

% Normalizing centers of the cluster
centers_of_cluster=centers_of_cluster./sqrt(sum(centers_of_cluster.^2,2));

point_index=zeros(h*h,1);

centers_of_cluster_new=centers_of_cluster;
centers_of_cluster_old=zeros(number_of_cluster,row_len);
epsi=0.01;
while  norm(centers_of_cluster_new-centers_of_cluster_old)>epsi
    centers_of_cluster_old=centers_of_cluster_new;
    
    %Clustering
    for i=1:h*h
        dist2point_btw_center=centers_of_cluster_old-V_norm(i,:);
        dist2point_btw_center=sum(dist2point_btw_center.^2,2);
        [min_num,min_index]=min(dist2point_btw_center);
        point_index(i)=min_index;
    end


    %Updating Centers
    for i=1:number_of_cluster
        indexes_of_cluster=find(point_index==i);
        %If there are no points assigned to center, center is updated randomly (and normalized)   
        if isempty(indexes_of_cluster)~=1
            %If there are less than "min_pixel" point assigned to center, center is updated randomly (and normalized) 
            if length(indexes_of_cluster)<min_pixel
                centers_of_cluster_new(i,:)=rand(1,row_len)-0.5;
                centers_of_cluster_new(i,:)=centers_of_cluster_new(i,:)./sqrt(sum(centers_of_cluster_new(i,:).^2));
            else
                centers_of_cluster_new(i,:)=sum(V_norm(indexes_of_cluster,:),1)./length(indexes_of_cluster);
            end
        else
            centers_of_cluster_new(i,:)=rand(1,row_len)-0.5;
            centers_of_cluster_new(i,:)=centers_of_cluster_new(i,:)./sqrt(sum(centers_of_cluster_new(i,:).^2));
        end
    end
    
    
    
end


```

```octave
clustered_matrix=reshape(point_index,64,64);

for i=1:number_of_cluster
    imag=zeros(64,64,3);
    dummy=zeros(64,64);
    dummy(clustered_matrix==i)=1;
    
    
    for j=1:3
        imag(:,:,j)=X_im(:,:,j).*dummy;
    end
    imag= cast(imag,"uint8"); 
   
    if mod(i,4)-1==0
        figure;
    end
    subplot(2,2,mod(mod(i,4)+3,4)+1);
    imshow(imag);
    title(['Cluster ' num2str(i)]);
end
```

```octave
hlf=number_of_cluster/2;
im_allin=zeros(64,64,3);
for i=1:number_of_cluster
    
    dummy=zeros(64,64);
    dummy(clustered_matrix==i)=1;
    
    im_allin(:,:,1)=( ones(64)*120*ceil(i/hlf) ) .* dummy + im_allin(:,:,1);
    im_allin(:,:,2)=( ones(64)*80* ceil((mod(mod(i,hlf)+hlf-1,hlf)+1)/3) ) .* dummy+im_allin(:,:,2);
    im_allin(:,:,3)=( ones(64)*80*mod((mod(mod(i,hlf)+hlf-1,hlf)+1),3)) .* dummy+im_allin(:,:,3);
   
   
end
im_allin= cast(im_allin,"uint8"); 
imshow(im_allin);
```
