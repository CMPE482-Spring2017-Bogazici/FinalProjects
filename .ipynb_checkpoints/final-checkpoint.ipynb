# 1- Project 1: Interpolation with Autoregressive (AR) Model


```python
## import useful libs 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from pandas import DataFrame
%matplotlib inline
```

Autoregressive model is a model where every element $X_t$ of the time series can be represented
by the weighted sum of the terms $X_{t−1}; X_{t−2}; \dots ; X_{t−M}$ plus a gaussian error term $\epsilon_t$.


$$
\begin{bmatrix}
    X_{M} \\
    X_{M+1}   \\
    \vdots  \\
    X_{N}  
\end{bmatrix}
=
\begin{bmatrix}
    X_{M-1} & X_{m-2} &\dots & X_{0}  \\
    X_{M} & X_{M-1} & \dots   & X_{1}\\
    \vdots & \vdots & \ddots  &  \vdots\\
    X_{N-1} & X_{N-2} & \dots   & X_{N-M} 
\end{bmatrix}
\begin{bmatrix}
    A_{1} \\
    A_{2}   \\
    \vdots  \\
    A_{M}  
\end{bmatrix}
+
\begin{bmatrix}
    \epsilon_{M} \\
    \epsilon_{M+1}   \\
    \vdots  \\
    \epsilon_{N}
\end{bmatrix}
$$

where we make the assumption $X_{M\dots N} \approx G_XA$

$$
\begin{bmatrix}
    \epsilon_{M} \\
    \epsilon_{M+1}   \\
    \vdots  \\
    \epsilon_{N-1}\\
    \epsilon_{N}
\end{bmatrix}
=
\begin{bmatrix}
    A^{T} & -1 & 0 & 0 & \dots  &  0\\
    0 & A^{T} & -1   & 0 &\dots  & 0\\
    \vdots & \ddots & \ddots  & \ddots & \ddots & \vdots \\
    0 & \dots & 0   & A^{T} & -1 & 0 \\
    0 & \dots & 0   & 0 & A^{T} & -1
\end{bmatrix}
\begin{bmatrix}
    X_{0} \\
    X_{1}   \\
    \vdots  \\
    X_{N-1} \\
    X_N
\end{bmatrix}
$$

where we make the assumption  $-G_AX^{ob} \approx G_AX^{mi}$

Our approach to solve this problem can be alternating least squares.

. Fix X

Compute

$A =  (G_X^TG_X)^{-1}(G_X^TG_X)$.

. Fix A

. Get rid of zeros in $X_{mi}$ and corresponding columns of $G_A$.


$X^{mi} = -(G_A^TG_A)^{-1}(G_A^TG_AX^{ob})$.


Update X with new $X^{mi}$ values

```python
#Fill nan values by 0
from collections import Counter
data2 = data.apply(lambda x: x.fillna(0,axis=0))
```

```python
'''
Construct Gx matrix
'''
def conver2Gx(y,M,N):
    Gx = np.zeros(shape=(N-M,M))
    for i in range(M,len(y)):
        Gx[i-4,:]=y[(i-4):i][::-1].T
    
    return  Gx  

'''
Construct Ga matrix
'''
def conver2Ga(A,M,N):
    Ga = np.zeros(shape=(N-M,N))
    for i in range(0,N-M):
        Ga[i,i:i+M+1]= np.append(A.T,-1)
    
    return Ga
```

```python
M = 4
N = len(data2['Passengers'])
Xob = data2['Passengers'].values.reshape(N,1)
Xmi = np.zeros(shape=(N,1))
missInd = np.where(Xob == 0)[0]
Xmi[missInd] = 1
missInd2 = np.where(Xmi == 0)[0]
# Construction X as Xob and Xmi
# In for loop solve for A and Xm by using least square
X = Xob+Xmi
for i in range(500):
    #Xmn
    y = X[M:]
    Gx = conver2Gx(X,M,N)
    A = np.linalg.lstsq(Gx,y)[0]
    Ga = conver2Ga(A,M,N)
    Xm = np.linalg.lstsq(np.delete(Ga,missInd2,1),-np.delete(Ga,missInd,1).dot(np.delete(Xob,missInd,0)))[0]
    X[missInd]= Xm
    if(i % 100 == 0):
        print(A)
```

 Assign new X to dataframe and plot. As we can see that new data does not seem to fit into missing values. If you try to plot first draft of the data, you can see that graphs don't overlap.

    This time we add 2 more terms to our model forming an affine function of time T. Coefficients of those terms are C and D. They are both appended to coefficient vector A.

$$
\begin{bmatrix}
    X_{M} \\
    X_{M+1}   \\
    \vdots  \\
    X_{N}  
\end{bmatrix}
=
\begin{bmatrix}
    X_{M-1} & X_{m-2} &\dots & X_{0} & M & 1\\
    X_{M} & X_{M-1} & \dots   & X_{1} & M+1 & 1\\
    \vdots & \vdots & \ddots  &  \vdots & \vdots & \vdots\\
    X_{N-1} & X_{N-2} & \dots   & X_{N-M}& N & 1 
\end{bmatrix}
\begin{bmatrix}
    A_{1} \\
    A_{2}   \\
    \vdots  \\
    A_{M} \\
    C \\
    D
\end{bmatrix}
+
\begin{bmatrix}
    \epsilon_{M} \\
    \epsilon_{M+1}   \\
    \vdots  \\
    \epsilon_{N}
\end{bmatrix}
$$


Then by using the representation given below we have this equation : $-(G_{A}X^{ob} + G_{F})  \approx G_AX^{mi} $ 

Where $ G_{F} = \begin{bmatrix}
    M & 1\\
    M+1& 1 \\
    \vdots&\vdots \\
    N-1 & 1\\
    N & 1
\end{bmatrix}
\begin{bmatrix}
    C\\
    D
\end{bmatrix}
$

$$\begin{bmatrix}
    \epsilon_{M} \\
    \epsilon_{M+1}   \\
    \vdots  \\
    \epsilon_{N-1}\\
    \epsilon_{N}
\end{bmatrix}
=
\begin{bmatrix}
    A^{T} & -1 & 0 & 0 & \dots  &  0\\
    0 & A^{T} & -1   & 0 &\dots  & 0\\
    \vdots & \ddots & \ddots  & \ddots & \ddots & \vdots \\
    0 & \dots & 0   & A^{T} & -1 & 0 \\
    0 & \dots & 0   & 0 & A^{T} & -1
\end{bmatrix}
\begin{bmatrix}
    X_{0} \\
    X_{1}   \\
    \vdots  \\
    X_{N-1} \\
    X_N
\end{bmatrix}
+
\begin{bmatrix}
    M & 1\\
    M+1& 1 \\
    \vdots&\vdots \\
    N-1 & 1\\
    N & 1
\end{bmatrix}
\begin{bmatrix}
    C\\
    D
\end{bmatrix}
$$

```python
data2 = data.apply(lambda x: x.fillna(0,axis=0))
Xab = data2['Passengers'].values.reshape(N,1)
Xmi = np.zeros(shape=(N,1))
missInd = np.where(Xab == 0)[0]
Xmi[missInd] = 1
missInd2 = np.where(Xmi == 0)[0]
X = Xab+Xmi
Ts = np.linspace(M,N-1,N-M).reshape(N-M,1)
Ones = np.ones(shape=(N-M,1))
# Constructing T matrix.
# In for loop we use least squares to calculate A and Xm
extraCols = np.concatenate((Ts,Ones),axis=1)
for i in range(500):
    # Xmn
    y = X[M:]
    Gx = conver2Gx(X,M,N)
    GxN = np.concatenate((Gx,extraCols),axis=1)
    A = np.linalg.lstsq(GxN,y)[0]
    Ga = conver2Ga(A[0:M],M,N)
    Xm = np.linalg.lstsq(Ga,-Ga.dot(Xab)-extraCols.dot(A[-2:]))[0]
    X[missInd]= Xm[missInd]
    if(i % 100 == 0):
        print(A)
```

# 2- Project 2: Image Segmentation with Spectral Clustering


```python
import scipy as sc
X = sc.misc.imread("stones_rgb.jpg")
```

```python
plt.imshow(X)
```

```python
# Variables are set to be used later

lamb = np.array([[0.299,0.587,0.114],[-0.14713,-0.28886, 0.436],[0.615,-0.51499,-0.10001]])
r = 2
sigma = 10.
H = X.shape[0]
N = H*H
W = np.zeros(shape=(N,N)) 
```

```python
# For each unique pixel-pixel pairs we calculate edge weights
# iff their x coordinates and y coordinates are close each other at least 'r' units. 

for i in range(N):
    for j in range(i+1,N):
        x_r = i%H
        x_c = j%H
        y_r = (i - x_r)//H
        y_c = (j - x_c)//H
        if i != j and np.absolute(x_r - x_c) <= r and np.absolute(y_r - y_c) <= r:
            W[i,j]=np.exp(-np.sum(lamb.dot(X[x_r, y_r]-X[x_c, y_c])**2)/(2*sigma**2))

# Since weight matrix is symmetric, 
#we just take the transpoze of it and maximum with its original entries
W = np.maximum(W,W.T)           
```

```python
# Creating and filling diagonal matrix
D = np.zeros(shape=(N,N))
sumCols = np.sum(W,axis=1)
D[range(N),range(N)] = sumCols
```

```python
# To get the inverse of D
# it is enough to take 1/sqrt of diagonal elements.
diagElms = np.where(np.diag(D))
d = D[diagElms,diagElms]
Dinv = np.zeros(shape=(N,N))
Dinv[diagElms,diagElms] = 1/np.sqrt(d)
```

```python
# Calculating laplacian matrix
L = Dinv.dot(W.dot(Dinv))
```

```python
# Starting iterative qr algorithm
# First 'K' eigenvector corresponding to largest eigenvalues are calculated.
K = 5
q,r = np.linalg.qr(np.random.rand(N, K))
for itr in range(2000):
    q,r = np.linalg.qr(L.dot(q))
    if(itr % 1000 == 0):
        print(q)
```

```python
# Calculate V
V = Dinv.dot(q)
```

```python
# We have K columns.
# For each column, 
#    we plot histogram 
#    we plot graymap of reshaped version 
#    we plot cluster 1
#    we plot cluster 2
# subplot(K,4,i) means : K rows , 4 columns , ith place for the plot.

for i in range(K):    
    plt.figure(figsize=(15,15))
    plt.subplot(K,4,i*4+1)
    plt.hist(V[:,i])
    plt.title(r'Histogram of $V_%d$'%(i+1))
    plt.subplot(K,4,i*4+2)
    plt.imshow(V[:,i].reshape(H, H),cmap='gray')
    plt.title(r'$V_%d$'%(i + 1))
    plt.subplot(K,4,i*4+3)
    C1 = X.copy()
    c1Ind = np.where(V[:,i].reshape(H, H) >= 0)
    C1[c1Ind] = 0
    plt.imshow(C1)
    plt.title('Cluster 1')
    plt.subplot(K,4,i*4+4)
    C2 = X.copy()
    c2Ind = np.where(V[:,i].reshape(H, H) < 0)
    C2[c2Ind] = 0
    plt.imshow(C2)
    plt.title('Cluster 2')
```

# Bonus Part

```python
# first of all, we normalize V matrix wrt its rows
normVec = np.linalg.norm(V, axis=1)
V_norm = V/normVec.reshape((V.shape[0],1))
```

```python
from sklearn.cluster import KMeans
# Clusters give us the labels as number for our matrix
clusters = KMeans(n_clusters=13, random_state=0).fit(V_norm).predict(V_norm)
```
